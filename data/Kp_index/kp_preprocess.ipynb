# Kp Index
Recolected from [here](https://wdc.kugi.kyoto-u.ac.jp/wdc/Sec3.html) and now preprocessed:

**raw data**: data\Kp_2015_to_2023.csv

**clean data**: data/Kp_index/data.csv


```python
with open('data.csv', 'w') as clean:
    with open('../Kp_2015_to_2023.csv', 'r') as raw:
        lines = raw.readlines()
        for line in lines:
            cleaned = line.replace('    ', ' ').replace('   ',' ').replace('  ', ' ').replace('+', '+ ').replace('-', '- ').replace('  ',' ').replace(' ',',')
            clean.write(cleaned)
            
```
